```python
import pandas as pd
from pandas.plotting import scatter_matrix, parallel_coordinates
import matplotlib.pylab as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('Application_Data.csv')
```

```python
print(df.info())
print(df.head())
```

```python
numerical_attributes = ['Applicant_ID', 'Owned_Car', 'Owned_Realty','Total_Children','Total_Income',
                        'Owned_Mobile_Phone','Owned_Work_Phone','Owned_Phone','Owned_Email', 'Total_Family_Members',
                        'Applicant_Age','Years_of_Working','Total_Bad_Debt','Total_Good_Debt','Status']
for att in numerical_attributes:
  plt.subplot(2,1,1)
  df[att].plot.hist()
  plt.subplot(2,1,2)
  df[att].plot.box(vert=False)
  plt.tight_layout()
  plt.savefig('{}.png'.format(att), dpi=600)
  plt.show() 
```

```python
categorical_attributes = ['Applicant_Gender', 'Income_Type', 'Education_Type', 'Family_Status',
                          'Housing_Type','Job_Title']
for att in categorical_attributes:
  df[att].value_counts().plot.barh()
  plt.title(att)
  plt.tight_layout()
  plt.savefig('{}.png'.format(att), dpi=600)
  plt.show()
```

```python
Applicant_Gender_possibilities = df.Applicant_Gender.unique()

box_sr = pd.Series('',index = Applicant_Gender_possibilities)

for poss in Applicant_Gender_possibilities:
  BM = df.Applicant_Gender == poss
  box_sr[poss] = df[BM]['Applicant_Age']

plt.boxplot(box_sr,vert=False)
plt.yticks([1,2],Applicant_Gender_possibilities)
plt.show() 
```

```python
Applicant_Gender_possibilities = df.Applicant_Gender.unique()


for poss in Applicant_Gender_possibilities:
  BM = df.Applicant_Gender == poss
  plt.hist(df[BM]['Applicant_Age'],
           histtype='step',label=poss)

plt.legend()
plt.show()
```

```python
Applicant_Gender_possibilities = df.Applicant_Gender.unique()

dataForBox_dic= {}

for poss in Applicant_Gender_possibilities:
  BM = df.Applicant_Gender == poss
  dataForBox_dic[poss] = df[BM]['Applicant_Age']

plt.subplot(2,1,1)
plt.boxplot(dataForBox_dic.values(),vert=False)
plt.yticks([1,2],Applicant_Gender_possibilities)
plt.show()

for poss in Applicant_Gender_possibilities:
  BM = df.Applicant_Gender == poss
  plt.hist(df[BM]['Applicant_Age'],histtype='step',label=poss)
  
plt.legend()
plt.tight_layout()
plt.show()
```

```python
contingency_tbl = pd.crosstab(df.Housing_Type, df.Applicant_Gender)
contingency_tbl
```

```python
probablity_tbl = contingency_tbl/ contingency_tbl.sum()
sns.heatmap(probablity_tbl, annot=True, center=0.5 ,cmap="Greys")
plt.show()
```

```python
contingency_tbl = pd.crosstab(df.Family_Status, df.Housing_Type)
probablity_tbl = contingency_tbl/ contingency_tbl.sum()
sns.heatmap(probablity_tbl, annot=True, center=0.5 ,cmap="Greys")
plt.show()
```
